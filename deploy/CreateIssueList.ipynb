```python
import numpy as np
import pandas as pd
```

```python
issues=pd.read_json("https://api.github.com/repos/neutronimaging/imagingsuite/issues?state=all&since=\"2018-08-01T10:00:00Z\"&per_page=200")
```

```python
closedissues=issues[(issues['state']=="closed") & (issues['pull_request'].isnull())]
closedissues[['number','title']]
```

# Print list for github markdown

```python
for index,issue in closedissues.iterrows() :
    print('- Issue',issue['number'], '- [',issue['title'],'](', issue['html_url'],')')
```

```python
committed=pd.DataFrame(columns=['name','date','message'])
for page in np.arange(1,10):
    url="https://api.github.com/repos/neutronimaging/imagingsuite/commits?state=all&page={}&since=\"2018-08-01T10:00:00Z\"&per_page=100".format(page)
    print(url)
    commits=pd.read_json(url)
    
    for index,commit in commits.iterrows() :
        c=commit["commit"]
        row={'name': c["committer"]["name"], 'date' : c["committer"]["date"], "message" : c["message"]}
        committed.loc[100*(page-1)+index]=row
    
    
```
